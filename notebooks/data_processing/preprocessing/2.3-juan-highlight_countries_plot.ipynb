```python
import matplotlib.pyplot as plt
import numpy as np

# Set the data
countries = ['Portugal', 'Germany', 'Austria']
r2_ml = [0.59, 0.49, 0.42]
r2_soa = [-16.98, 0.28, -3.57]

x = np.array([0, 1, 2])  # the label locations, closer together
width = 0.35  # the width of the bars

# Function to add labels inside the bars
def autolabel(rects, ax):
    for rect in rects:
        height = rect.get_height()
        if height > 0:  # If bar is above x-axis
            position = height / 2
        else:  # If bar is below x-axis
            position = height / 10  # Adjust this value to make the label appear above the bar
        ax.annotate('{:.2f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, position),
                    xytext=(0, 0),  # No vertical offset
                    textcoords="offset points",
                    ha='center', va='center', color='black', weight='bold')

# Create a figure with a specific size
fig, axs = plt.subplots(2, figsize=(10, 7))

# Create the first subplot for Machine-Learning based disaggregation
rects1 = axs[0].bar(x, r2_ml, width, color=['b', 'lightblue', 'cyan'])
axs[0].set_xticks(x)
axs[0].set_xticklabels(countries)
axs[0].set_ylabel('R2', fontsize=14)  # Increase the fontsize
axs[0].set_title('Machine-Learning based Spatial Disaggregation of Charging Stations for Electrical Vehicles', fontweight='bold')
autolabel(rects1, axs[0])  # Add labels

# Create the second subplot for State of the Art disaggregation
rects2 = axs[1].bar(x, r2_soa, width, color=['r', 'salmon', 'darkred'])
axs[1].set_xticks(x)
axs[1].set_xticklabels(countries)
axs[1].set_ylabel('R2', fontsize=14)  # Increase the fontsize
axs[1].set_title('State of the Art Spatial Disaggregation of Charging Stations for Electrical Vehicles', fontweight='bold')
autolabel(rects2, axs[1])  # Add labels

# Add some padding between subplots
fig.tight_layout(pad=3.0)

# Show the plot
plt.show()
```
