```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_r2_graph(data_path, variable_name):
    data = pd.read_excel(data_path)
    data['R2'] = pd.to_numeric(data['R2'].astype(str).str.replace(',', '.'), errors='coerce')
    countries = data['Country'].unique()
    fig, ax = plt.subplots(figsize=(20, 10))

    min_r2 = data['R2'].min()
    max_r2 = data['R2'].max()

    country_colors = {
        'Portugal': 'blue',
        'Germany': 'green',
        'Austria': 'red'
    }

    for country in countries:
        country_data = data[data['Country'].str.strip() == country]
        population_based = country_data[country_data['Disaggregation Method'].str.strip() == 'Population-based']
        model_based = country_data[country_data['Disaggregation Method'].str.strip() == 'Model-based']
        print(f"{country} - Model-based X: {model_based['Iteration'].values}, Y: {model_based['R2'].values}")
        print(f"{country} - Population-based Y: {population_based['R2'].values}")
        ax.plot(model_based['Iteration'], model_based['R2'], marker='o', linestyle='-', label=f'{country} - Model-based', color=country_colors[country])
        ax.hlines(population_based['R2'].iloc[0], xmin=0, xmax=model_based['Iteration'].max(), linestyle='--', label=f'{country} - Population-based', color=country_colors[country])
        pop_r2 = population_based['R2'].iloc[0]
        ax.annotate(f'{pop_r2:.2f}', (model_based['Iteration'].max() - 0.5, pop_r2), textcoords="offset points", xytext=(0,5), ha='center', fontsize=12, color=country_colors[country])

    ax.set_xlabel('Iteration')
    ax.set_ylabel('R2')
    ax.legend()
    ax.set_title(f'{variable_name} spatial disaggregation results - XGBoost - Self Supervised')
    ax.set_xlim(0.9, 5.1)
    ax.set_ylim(min_r2 - 1, max_r2 + 1)
    ax.set_yscale('symlog')
    plt.show()

cwd = os.getcwd()
DATA_PATH = os.path.join(cwd, '..', '..', '..', 'data', 'output', 'spatial_disaggregation', 'results', 'charging_stations_result.xlsx')
variable_name = "Charging Stations"  # or "Train Stations" depending on the dataset
plot_r2_graph(DATA_PATH, variable_name)
```
